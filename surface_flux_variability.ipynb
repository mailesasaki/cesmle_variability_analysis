```python
# Load modules

import xarray as xr
import pandas as pd
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import zarr
import gcsfs
```

# Control Surface Radiation

```python
control_hfls = xr.open_dataarray('/data/keeling/a/mailes2/cesmle_variability_analysis/processed_data/control_hfls.nc')
control_hfls.load()

control_hfss = xr.open_dataarray('/data/keeling/a/mailes2/cesmle_variability_analysis/processed_data/control_hfss.nc')
control_hfss.load()

control_rlds = xr.open_dataarray('/data/keeling/a/mailes2/cesmle_variability_analysis/processed_data/control_rlds.nc')
control_rlds.load()

control_rlus = xr.open_dataarray('/data/keeling/a/mailes2/cesmle_variability_analysis/processed_data/control_rlus.nc')
control_rlus.load()

control_rsds = xr.open_dataarray('/data/keeling/a/mailes2/cesmle_variability_analysis/processed_data/control_rsds.nc')
control_rsds.load()

control_rsus = xr.open_dataarray('/data/keeling/a/mailes2/cesmle_variability_analysis/processed_data/control_rsus.nc')
control_rsus.load()

vert_ds = xr.open_dataarray('/data/keeling/a/mailes2/cesmle_variability_analysis/processed_data/vert_ds.nc')
vert_ds.load()
```

```python
plt.rcParams['figure.figsize'] = [12,7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
control_hfls.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='Reds', vmin=0, vmax=7)
ax.coastlines(color='black', linewidth=1)
plt.title('Interdecadal Control Surface Upward Latent Heat Flux Variability (>25 yrs) - CMIP6 (N=37)')
```

```python
plt.rcParams['figure.figsize'] = [12,7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
control_hfss.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='Reds', vmin=0, vmax=7)
ax.coastlines(color='black', linewidth=1)
plt.title('Interdecadal Control Surface Upward Sensible Heat Flux Variability (>25 yrs) - CMIP6 (N=37)')
```

```python
plt.rcParams['figure.figsize'] = [12,7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
control_rlds.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='Reds', vmin=0, vmax=4)
ax.coastlines(color='black', linewidth=1)
plt.title('Interdecadal Control Surface Downwelling Longwave Flux Variability (>25 yrs) - CMIP6 (N=37)')
```

```python
plt.rcParams['figure.figsize'] = [12,7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
control_rlus.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='Reds', vmin=0, vmax=4)
ax.coastlines(color='black', linewidth=1)
plt.title('Interdecadal Control Surface Upwelling Longwave Flux Variability (>25 yrs) - CMIP6 (N=37)')
```

```python
plt.rcParams['figure.figsize'] = [12,7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
control_rsds.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='Reds', vmin=0, vmax=3)
ax.coastlines(color='black', linewidth=1)
plt.title('Interdecadal Control Surface Downwelling Shortwave Flux Variability (>25 yrs) - CMIP6 (N=37)')
```

```python
plt.rcParams['figure.figsize'] = [12,7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
control_rsus.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='Reds', vmin=0, vmax=3)
ax.coastlines(color='black', linewidth=1)
plt.title('Interdecadal Control Surface Upwelling Shortwave Flux Variability (>25 yrs) - CMIP6 (N=37)')
```

```python
control_qs = control_rlds + control_rsds - control_hfls - control_hfss - control_rlus - control_rsus
```

```python
plt.rcParams['figure.figsize'] = [12,7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
control_qs.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='RdBu_r', vmin=-2.5,vmax=2.5)
ax.coastlines(color='black', linewidth=1)
plt.title('Interdecadal Control Qs Variability (>25 yrs) - CMIP6 (N=37)')
```

```python
#with contours
plt.rcParams['figure.figsize'] = [12,7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
control_qs.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='RdBu_r', vmin=-2.5,vmax=2.5)
vert_ds.plot.contour(ax=ax, transform=ccrs.PlateCarree(), colors='#808080', levels=[-0.005,0.005])
ax.coastlines(color='black', linewidth=1)
plt.title('Interdecadal Control Qs Variability (>25 yrs) - CMIP6 (N=37)')
```
