```python
# Load all libraries

import xarray as xr
import pathlib
from os import listdir
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
from matplotlib import gridspec
import matplotlib
import xesmf as xe
import cartopy
```

# FLNS - Net Longwave Surface Flux

```python
files_FLNS = glob.glob('/data/keeling/a/cristi/a/data/LENS/LE_control/e.e11.E1850C5CN.f09_g16.001.cam.h0.FLNS.*')
print(files_FLNS)
print(len(files_FLNS))
```

```python
flns = xr.open_mfdataset(files_FLNS,combine="by_coords",use_cftime=True)
flns.load()
```

```python
new_lats=np.arange(-87,88,2)
new_lons=np.arange(1,359,2)
#new_time_index=pd.date_range(start='1920-02-16',freq='M', periods=1032)

flns_rein = flns.FLNS.interp(lat=new_lats, lon=new_lons)
#historical_newtime = historical_reinterp.assign_coords(time=new_time_index)
flns_rein
```

# FSNS - Net Shortwave Flux at the Surface

```python
files_FSNS = glob.glob('/data/keeling/a/cristi/a/data/LENS/LE_control/e.e11.E1850C5CN.f09_g16.001.cam.h0.FSNS.*')
print(files_FSNS)
print(len(files_FSNS))
```

# SHFLX - Shortwave Heat Flux

# LHFLX - Longwave Heat Flux

```python
files_LHFLX = glob.glob('/data/keeling/a/cristi/a/data/LENS/LE_control/e.e11.E1850C5CN.f09_g16.001.cam.h0.LHFLX.*')
print(files_LHFLX)
print(len(files_LHFLX))
```

```python
lhflx = xr.open_mfdataset(files_LHFLX,combine="by_coords",use_cftime=True)
lhflx.load()
```

```python
new_lats=np.arange(-87,88,2)
new_lons=np.arange(1,359,2)
#new_time_index=pd.date_range(start='1920-02-16',freq='M', periods=1032)

lhflx_rein = lhflx.LHFLX.interp(lat=new_lats, lon=new_lons)
#historical_newtime = historical_reinterp.assign_coords(time=new_time_index)
lhflx_rein
```

# Surface Fluxes Added Together

```python
surface = fsns_rein - flns_rein - lhflx_rein - shflx_rein
surface
```

```python
surface.to_netcdf('/data/keeling/a/mailes2/a/alt_compset_processed/CESM1_E_Surfacefluxes.nc')
```

```python
plt.rcParams['figure.figsize'] = [12, 7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
surface.mean('time').plot(ax=ax, transform=ccrs.PlateCarree(), cmap='Reds', vmin=-200, vmax=200)
plt.title('Surface Fluxes - Slab Ocean')
ax.coastlines(color='black', linewidth=1)
```
