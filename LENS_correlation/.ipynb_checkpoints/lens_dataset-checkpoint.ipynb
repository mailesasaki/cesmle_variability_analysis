# Catalog for CMIP6

```python
filepathlist = pd.read_csv('/data/keeling/a/cristi/a/esm_data/cmip6_lens/subdir_list.csv',\
                       header=None,names=['dirpaths']).values.tolist()

# Transform list of lists to list of strings
filepath=list(itertools.chain.from_iterable(filepathlist))
```

```python
# Create the Builder
cat_builder = Builder(
    # Directory of Organized CMIP6 files 
    paths=filepath)

# Build the Catalog
catalog = cat_builder.build(parsing_func=parse_cmip6)
```

```python
# Browse the catalog
catalog.df
```

```python
# Save the catalog as .csv 
catalog.df.to_csv('/data/keeling/a/cristi/a/esm_data/cmip6lens_catalog.csv', index=False)
```

# Create Dataset!

## CMIP6

```python
cat = pd.read_csv('/data/keeling/a/cristi/a/esm_data/cmip6lens_catalog.csv')
cat
```

## TAS

```python
new_lats=np.arange(-87,88,2)
new_lons=np.arange(1,359,2)
new_time_index=pd.date_range(start='1850-01-16',freq='M', periods=1980)

tas_list = []

for model in cat['source_id'].drop_duplicates():
    ds_version = []
    print(model)
    for version in cat.loc[(cat['variable_id']=='tas') & (cat['source_id']==model)]['member_id'].drop_duplicates():
        path = cat.loc[(cat['variable_id']=='tas') & 
                       (cat['source_id']==model) &
                       (cat['member_id']==version)]['path'].to_list()
        ds_v = xr.open_mfdataset(path,combine='by_coords',use_cftime=True)
        ds_version.append(ds_v)
    ds = xr.concat(ds_version, dim='version',coords='minimal', compat='override')
    ds_reinterp = ds.tas.interp(lat=new_lats, lon=new_lons)
    ds_time = ds_reinterp.assign_coords(time=new_time_index)
    ds_time['name'] = model
    tas_list.append(ds_time)
```

```python
tas_cmip6 = xr.concat(tas_list, dim='name',coords='minimal', compat='override')
tas_cmip6.load()
```

```python
tas_cmip6.to_netcdf('/data/keeling/a/mailes2/a/lens_processed/CMIP6_TAS_17mod_10mem.nc')
```
