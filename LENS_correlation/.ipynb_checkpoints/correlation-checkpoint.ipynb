```python
# Load libraries

import xarray as xr
import pathlib
from os import listdir
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Load data

```python
cesm = xr.open_dataarray('/data/keeling/a/mailes2/a/lens_processed/CESM_LENS_TAS.nc')
cesm.load()

cmip6 = xr.open_dataarray('/data/keeling/a/mailes2/a/lens_processed/CMIP6_TAS_17mod_10mem.nc')
cmip6.load()
```

# Compute Ensemble Means

```python
cesm_ensmean = cesm.mean(dim='ensemble_mem')
cesm_ensmean
```

```python
cmip6_ensmean = cmip6.mean(dim='version')
cmip6_ensmean
```

# Smooth to decadal time scales

```python
cesm_indiv_smoothed = cesm.rolling(time=120, center=True).mean('time')
cesm_indiv_smoothed
```

```python
cesm_smoothed = cesm_ensmean.rolling(time=120, center=True).mean('time')
cesm_smoothed
```

```python
cmip6_indiv_smoothed = cmip6.rolling(time=120, center=True).mean('time')
cmip6_indiv_smoothed
```

```python
cmip6_smoothed = cmip6_ensmean.rolling(time=120, center=True).mean('time')
cmip6_smoothed
```
