```python
# Load libraries

import xarray as xr
import pathlib
from os import listdir
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
```

# Load data

```python
vert_ds = xr.open_dataarray('/data/keeling/a/mailes2/cesmle_variability_analysis/processed_data/vert_ds.nc')
vert_ds.load()
```

```python
vert_cesm = xr.open_dataarray('/data/keeling/a/mailes2/a/processed_var_data/CESM1_LENS_omega500_histav.nc')
vert_cesm.load()
```

```python
cesm = xr.open_dataarray('/data/keeling/a/mailes2/a/lens_processed/CESM_LENS_TAS.nc')
cesm.load()

cmip6 = xr.open_dataarray('/data/keeling/a/mailes2/a/lens_processed/CMIP6_TAS_17mod_10mem.nc')
cmip6.load()

cesm2 = xr.open_dataarray('/data/keeling/a/mailes2/a/lens_processed/CESM2_LENS_TAS.nc')
cesm2.load()
```

```python
cesm
```

# CESM

## Compute Ensemble Means

```python
cesm_ensmean = cesm.mean(dim='ensemble_mem')
cesm_ensmean
```

## Smooth to decadal time scales

```python
cesm_indiv_smoothed = cesm.rolling(time=120, center=True).mean('time')
cesm_indiv_smoothed
```

```python
cesm_smoothed = cesm_ensmean.rolling(time=120, center=True).mean('time')
cesm_smoothed
```

## Correlate local temperature

```python
cesm_local_corr = xr.corr(cesm_indiv_smoothed, cesm_smoothed, dim='time')
```

```python
cesm_local_corr
```

```python
cesm_local_corr_mean = cesm_local_corr.mean(dim='ensemble_mem')
cesm_local_corr_mean
```

```python
# CESM1 Omega 500 Contours
plt.rcParams['figure.figsize'] = [12, 7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
cesm_local_corr_mean.plot(ax=ax, transform=ccrs.PlateCarree(), vmin=0.2, vmax=1, cmap='jet')
ax.coastlines(color='black', linewidth=1)
vert_cesm.plot.contour(ax=ax, transform=ccrs.PlateCarree(), colors='#808080', levels=[-0.005,0.005])
plt.title('CESM1 TAS Correlation - Local temperature')
```

```python
# CESM1 Omega 500 Contours
plt.rcParams['figure.figsize'] = [12, 7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
cesm_local_corr_mean.plot(ax=ax, transform=ccrs.PlateCarree(), vmin=0.3, vmax=1.3, cmap='RdBu_r')
ax.coastlines(color='black', linewidth=1)
#vert_cesm.plot.contour(ax=ax, transform=ccrs.PlateCarree(), colors='#808080', levels=[-0.005,0.005])
plt.title('CESM1 TAS Correlation - Local temperature')
```

# Correlate global temperature

```python
cesm_global_corr = xr.corr(cesm_indiv_smoothed, cesm_smoothed.mean('lat').mean('lon'), dim='time')
```

```python
cesm_global_corr
```

```python
cesm_global_corr_mean = cesm_global_corr.mean('ensemble_mem')
```

```python
plt.rcParams['figure.figsize'] = [12, 7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
cesm_global_corr_mean.plot(ax=ax, transform=ccrs.PlateCarree(), vmin=0.2, vmax=1, cmap='jet')
ax.coastlines(color='black', linewidth=1)
vert_ds.plot.contour(ax=ax, transform=ccrs.PlateCarree(), colors='#808080', levels=[-0.005,0.005])
plt.title('CESM1 TAS Correlation - Global temperature')
```

```python
# CESM1 Omega 500 Contours
plt.rcParams['figure.figsize'] = [12, 7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
cesm_global_corr_mean.plot(ax=ax, transform=ccrs.PlateCarree(), vmin=0.2, vmax=1, cmap='jet')
ax.coastlines(color='black', linewidth=1)
vert_cesm.plot.contour(ax=ax, transform=ccrs.PlateCarree(), colors='#808080', levels=[-0.005,0.005])
plt.title('CESM1 TAS Correlation - Global temperature')
```

```python
# CESM1 Omega 500 Contours
plt.rcParams['figure.figsize'] = [12, 7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
cesm_global_corr_mean.plot(ax=ax, transform=ccrs.PlateCarree(), vmin=0.3, vmax=1.3, cmap='RdBu_r')
ax.coastlines(color='black', linewidth=1)
#vert_cesm.plot.contour(ax=ax, transform=ccrs.PlateCarree(), colors='#808080', levels=[-0.005,0.005])
plt.title('CESM1 TAS Correlation - Global temperature')
```

# Smooth to 5 year timescales

```python
cesm_indiv_smoothed_5yr = cesm.rolling(time=60, center=True).mean('time')
cesm_smoothed_5yr = cesm_ensmean.rolling(time=60, center=True).mean('time')
cesm_local_corr_5yr = xr.corr(cesm_indiv_smoothed_5yr, cesm_smoothed_5yr, dim='time')
cesm_local_corr_mean_5yr = cesm_local_corr_5yr.mean(dim='ensemble_mem')
cesm_local_corr_mean_5yr
```

```python
# CESM1 Omega 500 Contours
plt.rcParams['figure.figsize'] = [12, 7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
cesm_local_corr_mean_5yr.plot(ax=ax, transform=ccrs.PlateCarree(), vmin=0.2, vmax=1, cmap='jet')
ax.coastlines(color='black', linewidth=1)
vert_cesm.plot.contour(ax=ax, transform=ccrs.PlateCarree(), colors='#808080', levels=[-0.005,0.005])
plt.title('CESM1 TAS Correlation - Local temperature - 5 year smoothing')
```

```python
cesm_global_corr_5yr = xr.corr(cesm_indiv_smoothed_5yr, cesm_smoothed_5yr.mean('lat').mean('lon'), dim='time')
cesm_global_corr_mean_5yr = cesm_global_corr_5yr.mean('ensemble_mem')
```

```python
# CESM1 Omega 500 Contours
plt.rcParams['figure.figsize'] = [12, 7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
cesm_global_corr_mean_5yr.plot(ax=ax, transform=ccrs.PlateCarree(), vmin=0.2, vmax=1, cmap='jet')
ax.coastlines(color='black', linewidth=1)
vert_cesm.plot.contour(ax=ax, transform=ccrs.PlateCarree(), colors='#808080', levels=[-0.005,0.005])
plt.title('CESM1 TAS Correlation - Global temperature - 5 year smoothing')
```
