```python
# Load all libraries

import xarray as xr
import pathlib
from os import listdir
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import cartopy.crs as ccrs
from matplotlib import gridspec
import matplotlib
import xesmf as xe
import cartopy
```

# FLNS - Net Longwave Flux at the Surface

```python
files_F_FLNS = glob.glob('/data/keeling/a/cristi/a/data/LENS/LE_control/f.e11.F1850C5CN.f09_f09.001.cam.h0.FLNS.*')
print(files_F_FLNS)
print(len(files_F_FLNS))
```

```python
flns = xr.open_mfdataset(files_F_FLNS,combine="by_coords",use_cftime=True)
flns.load()
```

```python
new_lats=np.arange(-87,88,2)
new_lons=np.arange(1,359,2)
#new_time_index=pd.date_range(start='1920-02-16',freq='M', periods=1032)

flns_rein = flns.FLNS.interp(lat=new_lats, lon=new_lons)
#historical_newtime = historical_reinterp.assign_coords(time=new_time_index)
flns_rein
```

```python
files_F_SHFLX = glob.glob('/data/keeling/a/cristi/a/data/LENS/LE_control/f.e11.F1850C5CN.f09_f09.001.cam.h0.SHFLX.*')
print(files_F_SHFLX)
print(len(files_F_SHFLX))
```

```python
new_lats=np.arange(-87,88,2)
new_lons=np.arange(1,359,2)
#new_time_index=pd.date_range(start='1920-02-16',freq='M', periods=1032)

shflx_rein = shflx.SHFLX.interp(lat=new_lats, lon=new_lons)
#historical_newtime = historical_reinterp.assign_coords(time=new_time_index)
shflx_rein
```
