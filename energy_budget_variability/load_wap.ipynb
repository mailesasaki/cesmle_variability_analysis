```python
# Load modules

import xarray as xr
import pandas as pd
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import zarr
import gcsfs
import xesmf as xe
from scipy.interpolate import griddata
from xmip.preprocessing import rename_cmip6
from matplotlib import gridspec
```

# piControl

```python
# Head to google catalog to get some vertical motion data

# Read in data catalog

df = pd.read_csv('https://storage.googleapis.com/cmip6/cmip6-zarr-consolidated-stores.csv')
df.head()
```

# piControl

```python
df_wap_cont = df.query("activity_id=='CMIP' & table_id == 'Amon' & variable_id == 'wap' & experiment_id == 'piControl' & member_id=='r1i1p1f1'")
#df_sst_control = df_sst_control.drop_duplicates('source_id').head(44)
df_wap_cont = df_wap_cont.drop_duplicates('source_id')
df_wap_cont = df_wap_cont[~df_wap_cont['source_id'].isin(['MCM-UA-1-0','GISS-E2-1-G-CC','NorESM1-F','IITM-ESM',
                                                                  'GISS-E2-2-G','AWI-ESM-1-1-LR','EC-Earth3-LR',
                                                                  'E3SM-1-1-ECA','MIROC-ES2H','ICON-ESM-LR',
                                                                  'KIOST-ESM','NESM3','KACE-1-0-G','GISS-E2-1-H','MPI-ESM-1-2-HAM',
                                                                  'HadGEM3-GC31-LL','EC-Earth3-Veg','CAS-ESM2-0','FIO-ESM-2-0',
                                                         'IPSL-CM6A-LR','E3SM-1-1'])]
df_wap_cont
```

```python
#this only needs to be created once
# From Ryan Abernathy's tutorial - Way 2
gcs = gcsfs.GCSFileSystem(token='anon')
fs = gcsfs.GCSFileSystem(token='anon', access='read_only')

# Regridding
new_lats=np.arange(-87,88,2)
new_lons=np.arange(1,359,2)

#time index to assign model times to
new_time_index=pd.date_range(start='1851-01-01',freq='M',periods=3000) #250 years
new_time_index
    
dset_list_wap = []
for zstore in df_wap_cont.zstore.unique():
    # make a nice concise name from the zstore value
    name = zstore.split('gs://cmip6/CMIP6/')[1].replace('/','.')[:-1]
    name = zstore.split('gs://cmip6/CMIP6/CMIP/')[1].replace('/','.')[:-1]
    name_list = name.split('.')
    name = name_list[1]
    print(name)
    ds = xr.open_zarr(fs.get_mapper(zstore),consolidated=True)
    ds = ds.sel(plev=50000, method='nearest')
    print(ds.time)
    ds_sel = ds.isel(time=slice(len(ds.time)-3000,len(ds.time)))
    print(ds_sel.time.head(1).values)
    print(ds_sel.time.tail(1).values)
    ds_wap_interp = ds_sel.assign(time=new_time_index).interp(lat=new_lats,lon=new_lons)
    ds_wap = ds_wap_interp.wap
    ds_wap['name'] = name # Creating a new dimension for each ensemble member
    dset_list_wap.append(ds_wap)
```

```python
wap_control = xr.concat(dset_list_wap, dim='name', coords='minimal', compat='override')
wap_control.load()
```

```python
wap_control.to_netcdf('/data/keeling/a/mailes2/a/processed_var_data/250yeardatasets/CMIP_wap_37model_piControl_2x2_last250.nc')
```

```python
fig = plt.figure(figsize=(20,20))
gs = gridspec.GridSpec(1, 2)
gs.update(wspace=0.1)

i = 1

for model in wap_control['name'].values:
    ax = fig.add_subplot(8,5,i, projection=ccrs.PlateCarree(central_longitude=180))
    wap_control.sel(name=model).mean('time').plot(ax=ax, transform=ccrs.PlateCarree(), vmin=-0.2, vmax=0.2, cmap='magma')
    ax.coastlines(color='black', linewidth=1)
    plt.title(model)
    i += 1
plt.tight_layout()
plt.suptitle('Mean WAP', fontsize=20, y=0.99)
```
