```python
# Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
from matplotlib import gridspec
import matplotlib
```

```python
mlotst = xr.open_dataarray('/data/keeling/a/mailes2/a/processed_var_data/250yeardatasets/CMIP_mlotst_piControl_250yr.nc')
mlotst.load()
```

```python
mlotst_yr = mlotst.rolling(time=12, center=True).mean('time')
```

```python
plt.rcParams['figure.figsize'] = [12, 7]
ax = plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
mlotst.mean('time').mean('name').plot(ax=ax, transform=ccrs.PlateCarree(), vmin=0, vmax=150, cmap='Reds')
ax.coastlines(color='black', linewidth=1)
plt.title('piControl MLOTST monthly mean (N=15)')
```
