```python
# Load libraries

import xarray
import pathlib
from os import listdir
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Look at an isolated dataset

data1 = xarray.open_dataset('/data/keeling/a/cristi/a/data/LENS/LE_hist/b.e11.B20TRC5CNBDRD.f09_g16.001.cam.h0.FLNS.185001-200512.nc')
data1

# This looks like a dataset that specifically has net longwave flux at surface
# In my notes, it says I should pull up TREFHT - temperature to reference height files
```

```python
# Find only the TREFHT files

files = glob.glob('/data/keeling/a/cristi/a/data/LENS/LE_hist/*.TREFHT.*')
print(files)
```

## Single dataset

```python
# Taking one dataset

dataset = xarray.open_dataset(files[0])
xarray.decode_cf(dataset)
global_mean = dataset.TREFHT.mean(("lat", "lon"))
print(type(global_mean))
global_mean.plot(x='time')
plt.title('TREFHT from 1920 to 2006')
dataset.TREFHT
```

```python
# One dataset - Getting anomalies
# https://datalab.marine.rutgers.edu/2020/11/introduction-to-python-data-analysis/

# Grabbing the 1961 to 1990 dataset

means_in_period = global_mean.loc["1961-01-01 00:00:00":"1990-01-01 00:00:00"]
means_in_period
```

```python
# One dataset - Anomaly calculation

#Making into Pandas dataframe

#Base period
df_base = means_in_period.to_dataframe()
datetimeindex_base = df_base.index.to_datetimeindex()
df_base.index = datetimeindex_base

#Whole timeline
df = global_mean.to_dataframe()
datetimeindex = df.index.to_datetimeindex()
df.index = datetimeindex
df
```

```python
# One dataset - anomalies from base period

climat_mean = df_base.groupby(df_base.index.month).mean()

anom_df = df.copy()
for mon in range(1,13):
    new_df = df.loc[(df.index.month==mon)]
    new_clim = climat_mean.loc[(climat_mean.index==mon, 'TREFHT')]
    val = new_clim.iloc[0]
    anom_df[(df.index.month==mon)] = new_df.sub(val)
```

```python
# One dataset - plot

anom_df.plot()
plt.title('Monthly anomaly in 1920-2006 from 1961-90 baseline')
plt.ylabel('TREFHT anomaly')
plt.xlabel('Year')
```

## For all data in the directory

```python
# Ensemble - Processing all of the datasets

LENS_list = []
ensem_num = 0
for file in files:
    model_run = xarray.open_dataset(file)
    xarray.decode_cf(model_run)
    TREFHT_data = model_run.TREFHT
    TREFHT_data['ensemble_member'] = ensem_num # Creating a new dimension for each ensemble member
    ensem_num += 1
    LENS_list.append(TREFHT_data)
LENS = xarray.concat(LENS_list, dim='ensemble_member')
LENS
```

```python
# Ensemble - Grabbing 1961-1990 base period
ensemble_global_means = LENS.mean(("lat", "lon"))
means_in_period_ensemble = ensemble_global_means.loc[dict(time=slice("1961-01-01 00:00:00", "1990-01-01 00:00:00"))]
means_in_period_ensemble
```

```python
# Ensemble 

#Making into Pandas dataframe

#Base period
ensem_base_df = means_in_period_ensemble.to_dataframe()
ensem_base_df.index = ensem_base_df.index.set_levels(
    ensem_base_df.index.levels[1].to_datetimeindex(), 
    level=1
)

print(ensem_base_df)
#Whole timeline
ensem_df = ensemble_global_means.to_dataframe()
ensem_df.index = ensem_df.index.set_levels(
    ensem_df.index.levels[1].to_datetimeindex(), 
    level=1
)
```

```python
# Ensemble - Calculating anomalies

climat_mean_ensem = ensem_base_df.groupby(ensem_base_df.index.get_level_values('time').month).mean()

ensem_anom_df = ensem_df.copy()
for mon in range(1,13):
    new_df = ensem_df.loc[(ensem_df.index.get_level_values('time').month==mon)]
    new_clim = climat_mean_ensem.loc[(climat_mean_ensem.index==mon, 'TREFHT')]
    val = new_clim.iloc[0]
    ensem_anom_df[(ensem_df.index.get_level_values('time').month==mon)] = new_df.sub(val)
ensem_anom_df
```

```python
# Mean of the ensemble

mean_ensem_anom = ensem_anom_df.groupby(level='time').mean()
mean_ensem_anom
```

```python
# Ensemble - plot

fig, ax = plt.subplots()

for ensem_mem in ensem_anom_df.index.levels[0]:
    ensem_anom_df.loc[ensem_mem,:].plot.line(ax=ax, color='gray', legend=False)
ensem_anom_df.loc[2,:].plot.line(ax=ax, color='black', legend=False)
mean_ensem_anom.loc['1920-01-01':].plot.line(ax=ax, color='red', legend=False)

plt.title('Monthly anomaly in 1850-2006 from 1961-90 baseline')
plt.ylabel('TREFHT anomaly')
plt.xlabel('Year')
```
