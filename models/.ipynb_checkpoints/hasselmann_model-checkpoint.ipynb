# Import packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import matplotlib
```

# Set parameters

```python
density = 1026 #Water density - kg/m3
c_p = 3850 #Specific heat of salt water - J/kgC
seconds = 7*24*60*60 #Time in a week - seconds
```

```python
mlotst_equator = 21.95173968
```

```python
def white_noise(sigma):
    """
    Define white noise using a normal distribution
    
    """
    mean = 0
    std = sigma
    
    return np.random.normal(mean, std)
```

```python
def integrate(params, years):
    """
    Integrating the white noise model 
    
    C dT/dt = -lambda T + F
    
    Inputs:
        params
            lambda = Atmospheric feedback
            sigma  = Variance of atmospheric forcing
            
    Outputs:
        T - Temperature [array]
        
    """
    lambda_ = params['lambda'] # Atmospheric feedback
    sigma = params['sigma']   # Variance of atmospheric forcing
    
    Nt = 52*years # Number of timesteps - weekly [weeks]
    dt = 60*60*24*7 # Timestep - week [s]
    
    np.random.seed(0)
    
    T = np.zeros(Nt+1) # Temperature array over time [K]
    
    for t in range(1,Nt):
        T[t+1] = T[t] + (dt/(c_p*density*mlotst_equator))*(-lambda_*T[t] + white_noise(sigma))
        
    return T    
```

```python
def plot(params, years):
    fig, ax = plt.subplots()
    ax.plot(np.arange(years*52+1), T)
    ax.set_xticks(np.arange(0,years*52+1,52*200), np.arange(0,years+1,200))
    ax.set_title((r'$\sigma$: ' + str(params['sigma']) + r'  $\lambda$:' + str(params['lambda'])))
    ax.set_xlabel('Time [years]')
    ax.set_ylabel('Temperature Anomaly [K]')
    fig.show()
```

# Running the model

```python
font = {'size'   : 15,
        'weight' : 'normal'}

matplotlib.rc('font', **font)
```

```python
params = {}
params['lambda'] = 0.01
params['sigma'] = 0.1
years = 1000
plt.rcParams['figure.figsize'] = [8, 5]

T = integrate(params, 1000)
plot(params, years)
plt.ylim(-0.15,0.15)
```
