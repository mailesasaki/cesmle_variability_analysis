```python
# Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
from matplotlib import gridspec
import matplotlib
from scipy.optimize import curve_fit
```

```python
ts_cont = xr.open_dataarray('/data/keeling/a/mailes2/a/finalized_var_data/CMIP_ts_39model_piControl_2x2_last250.nc')
ts_cont.load()

ocean_cont = xr.open_dataarray('/data/keeling/a/mailes2/a/finalized_var_data/CMIP_ocean_21mod_piControl_2x2_last250_1.nc')
ocean_cont.load()

surface_cont = xr.open_dataarray('/data/keeling/a/mailes2/a/finalized_var_data/CMIP_surface_21mod_piControl_2x2_last250.nc')
surface_cont.load()

toa_cont = xr.open_dataarray('/data/keeling/a/mailes2/a/finalized_var_data/CMIP_toa_25mod_piControl_2x2_last250.nc')
toa_cont.load()

horiz_cont = xr.open_dataarray('/data/keeling/a/mailes2/a/finalized_var_data/CMIP_horiz_20mod_piControl_2x2_last250.nc')
horiz_cont.load()

landmask = xr.open_dataarray('/data/keeling/a/mailes2/a/finalized_var_data/CMIP_landmask_31model.nc')
landmask.load()

omega_cont = xr.open_dataarray('/data/keeling/a/mailes2/a/finalized_var_data/CMIP_wap_37model_piControl_2x2_last250.nc')
omega_cont.load()
```

```python
models = ['ACCESS-CM2','BCC-CSM2-MR','BCC-ESM1','CAMS-CSM1-0','CanESM5','CESM2','CESM2-WACCM','CESM2-WACCM-FV2','CMCC-CM2-SR5',
          'FGOALS-g3','GFDL-CM4','GFDL-ESM4','GISS-E2-1-G','HadGEM3-GC31-MM','MIROC6','MPI-ESM1-2-HR',
          'MPI-ESM1-2-LR','MRI-ESM2-0','NorCPM1']

ts_cont_trunc = ts_cont.sel(name=models)
ocean_cont_trunc = ocean_cont.sel(name=models)
surface_cont_trunc = surface_cont.sel(name=models)
toa_cont_trunc = toa_cont.sel(name=models)
horiz_cont_trunc = horiz_cont.sel(name=models)
```

```python
# Calculating Anomalies

# Remove seasonality
climatology_sst = ts_cont_trunc.groupby('time.month').mean('time')
T_anomalies = ts_cont_trunc.groupby('time.month') - climatology_sst
T_minusone_prime = T_anomalies.isel(time=slice(0,2999))
T_prime = T_anomalies.shift(time=-1).isel(time=slice(0,2999))

climatology_Qs = surface_cont_trunc.groupby('time.month').mean('time')
Qs_anomalies = surface_cont_trunc.groupby('time.month') - climatology_Qs
Qs_prime = Qs_anomalies.shift(time=-1).isel(time=slice(0,2999))
            
climatology_Qo = ocean_cont_trunc.groupby('time.month').mean('time')
Qo_anomalies = ocean_cont_trunc.groupby('time.month') - climatology_Qo
Qo_prime = Qo_anomalies.shift(time=-1).isel(time=slice(0,2999))

climatology_Qt = toa_cont_trunc.groupby('time.month').mean('time')
Qt_anomalies = toa_cont_trunc.groupby('time.month') - climatology_Qt
Qt_prime = Qt_anomalies.shift(time=-1).isel(time=slice(0,2999))

climatology_Qh = horiz_cont_trunc.groupby('time.month').mean('time')
Qh_anomalies = horiz_cont_trunc.groupby('time.month') - climatology_Qh
Qh_prime = Qh_anomalies.shift(time=-1).isel(time=slice(0,2999))
```

# Figure 5

```python
# Surface fluxes
lambda_s_num = Qs_prime*T_minusone_prime
lambda_s_den = T_prime*T_minusone_prime
lambda_s = -lambda_s_num.mean('time')/lambda_s_den.mean('time')

# Ocean fluxes
lambda_o_num = Qo_prime*T_minusone_prime
lambda_o = -lambda_o_num.mean('time')/lambda_s_den.mean('time')

# TOA fluxes
lambda_t_num = Qt_prime*T_minusone_prime
lambda_t = -lambda_t_num.mean('time')/lambda_s_den.mean('time')

# Horiz. fluxes
lambda_h_num = Qh_prime*T_minusone_prime
lambda_h = -lambda_h_num.mean('time')/lambda_s_den.mean('time')
```

```python
fig = plt.figure(figsize=(35, 15))
gs = gridspec.GridSpec(2, 2, width_ratios=[2,2])
gs.update(hspace=0, wspace=0.1)
#top=300
#bottom=270
font = {'size'   : 30,
        'weight' : 'normal'}

matplotlib.rc('font', **font)

vmin = -30
vmax = 30

ax1 = fig.add_subplot(gs[0], projection=ccrs.PlateCarree(central_longitude=180))
lambda_s.mean('name').plot(ax=ax1, transform=ccrs.PlateCarree(), vmin=vmin, vmax=vmax, 
                           cmap='RdBu_r', cbar_kwargs = {"shrink": 0.7,
                                                         "label": r'$\lambda_s$ [$W/m^2K$]'})
ax1.coastlines(color='black', linewidth=1)
#ax1.gridlines(draw_labels=True)
#vert_ds.plot.contour(ax=ax1, transform=ccrs.PlateCarree(), colors='black', levels=[-0.005,0.005])
ax1.set_extent([0,360,-60,60], crs=ccrs.PlateCarree())
plt.title(r'$\lambda_s$ (N=19)')

ax2 = fig.add_subplot(gs[1], projection=ccrs.PlateCarree(central_longitude=180))
lambda_o.mean('name').plot(ax=ax2, transform=ccrs.PlateCarree(), vmin=vmin, vmax=vmax,
                               cmap='RdBu_r', cbar_kwargs = {"shrink": 0.7,
                                                             "label": r'$\lambda_o$ [$W/m^2K$]'})
ax2.coastlines(color='black', linewidth=1)
#vert_ds.plot.contour(ax=ax2, transform=ccrs.PlateCarree(), colors='black', levels=[-0.005,0.005])
ax2.set_extent([0,360,-60,60], crs=ccrs.PlateCarree())
plt.title(r'$\lambda_o$ (N=19)')

ax3 = fig.add_subplot(gs[2], projection=ccrs.PlateCarree(central_longitude=180))
lambda_t.mean('name').plot(ax=ax3, transform=ccrs.PlateCarree(), vmin=-5, vmax=5,
                           cmap='RdBu_r', cbar_kwargs = {"shrink": 0.7,
                                                         "label": r'$\lambda_t$ [$W/m^2K$]'})
ax3.coastlines(color='black', linewidth=1)
ax3.set_extent([0,360,-60,60], crs=ccrs.PlateCarree())
#vert_ds.plot.contour(ax=ax3, transform=ccrs.PlateCarree(), colors='black', levels=[-0.005,0.005])
plt.title(r'$\lambda_t$ (N=19)')

ax4 = fig.add_subplot(gs[3], projection=ccrs.PlateCarree(central_longitude=180))
lambda_h.mean('name').plot(ax=ax4, transform=ccrs.PlateCarree(), vmin=vmin, vmax=vmax,
                           cmap='RdBu_r', cbar_kwargs = {"shrink": 0.7,
                                                         "label": r'$\lambda_h$ [$W/m^2K$]'})
ax4.coastlines(color='black', linewidth=1)
ax4.set_extent([0,360,-60,60], crs=ccrs.PlateCarree())
#vert_ds.plot.contour(ax=ax4, transform=ccrs.PlateCarree(), colors='black', levels=[-0.005,0.005])
plt.title(r'$\lambda_h$ (N=19)')
```

```python
# Calculating Anomalies

# Remove seasonality
climatology_sst = ts_cont_trunc.groupby('time.month').mean('time')
T_anomalies = ts_cont_trunc.groupby('time.month') - climatology_sst
T_minusone_prime = T_anomalies.isel(time=slice(0,2999))
T_prime = T_anomalies.shift(time=-1).isel(time=slice(0,2999))

climatology_Qs = surface_cont_trunc.groupby('time.month').mean('time')
Qs_anomalies = surface_cont_trunc.groupby('time.month') - climatology_Qs
Qs_prime = Qs_anomalies.shift(time=-1).isel(time=slice(0,2999))
            
climatology_Qo = ocean_cont_trunc.groupby('time.month').mean('time')
Qo_anomalies = ocean_cont_trunc.groupby('time.month') - climatology_Qo
Qo_prime = Qo_anomalies.shift(time=-1).isel(time=slice(0,2999))
```

```python
# Surface fluxes
lambda_s_num = Qs_prime*T_minusone_prime
lambda_s_den = T_prime*T_minusone_prime
lambda_s = -lambda_s_num.mean('time')/lambda_s_den.mean('time')

# Ocean fluxes
lambda_o_num = Qo_prime*T_minusone_prime
lambda_o = -lambda_o_num.mean('time')/lambda_s_den.mean('time')
```

```python
F_o = Qo_prime+ lambda_o*T_prime

F_a = Qs_prime + lambda_s*T_prime

F_gen = F_o + F_a

F_std = F_gen.std('time')
```

```python
lambda_gen = lambda_s + lambda_o
```

```python
fig = plt.figure(figsize=(35, 15))
gs = gridspec.GridSpec(2, 2, width_ratios=[2,2])
gs.update(wspace=0)
#top=300
#bottom=270
font = {'size'   : 30,
        'weight' : 'normal'}

matplotlib.rc('font', **font)

vmin = -30
vmax = 30

ax1 = fig.add_subplot(gs[0], projection=ccrs.PlateCarree(central_longitude=180))
lambda_plot = lambda_gen.mean('name').plot(ax=ax1, transform=ccrs.PlateCarree(), vmin=vmin, vmax=vmax, add_colorbar=False,
                           cmap='RdBu_r', #cbar_kwargs = {"shrink": 0.7,
                                          #               "label": r'$\lambda$ [$W/m^2K$]',
                                          #               "orientation": 'horizontal'}
                                          )
ax1.coastlines(color='black', linewidth=1)
#ax1.gridlines(draw_labels=True)
#vert_ds.plot.contour(ax=ax1, transform=ccrs.PlateCarree(), colors='black', levels=[-0.005,0.005])
cb1 = plt.colorbar(lambda_plot, orientation="horizontal", pad=0.05, aspect=25, shrink=0.9, extend='max')
cb1.set_label(label=r'$\lambda$ [$W/m^2K$]')
ax1.set_extent([0,360,-60,60], crs=ccrs.PlateCarree())
plt.title(r'$\lambda$ (N=19)')

ax2 = fig.add_subplot(gs[1], projection=ccrs.PlateCarree(central_longitude=180))
f_plot = F_std.mean('name').plot(ax=ax2, transform=ccrs.PlateCarree(), vmin=10, vmax=10**2, cmap='plasma_r', add_colorbar=False,
                                  norm=matplotlib.colors.LogNorm(), #cbar_kwargs = {'shrink': 0.7,
                                                                    #               'ticks': [10e1],
                                                                    #               "label": r'$\sigma_F$ [$W/m^2$]',
                                                                    #               "orientation": 'horizontal'}
                                )
cb = plt.colorbar(f_plot, orientation="horizontal", pad=0.05, aspect=25, shrink=0.9, extend='both', 
                  ticks=[10**1, 2*10**1, 4*10**1, 6*10**1, 8*10**1,10**2])
cb.ax.minorticks_off()
#cb.ax.set_xticklabels([1e1, 2e1, 4e1, 6e1, 8e1,10e1])
#cb.set_ticklabels([r'$10^1$', r'$2x10^1$', r'$4x10^1$', r'$6x10^1$', r'$8x10^1$', r'$10^2$'])
ax2.coastlines(color='black', linewidth=1)
cb.set_label(label=r'$\sigma_F$ [$W/m^2$]')
#vert_ds.plot.contour(ax=ax2, transform=ccrs.PlateCarree(), colors='black', levels=[-0.005,0.005])
ax2.set_extent([0,360,-60,60], crs=ccrs.PlateCarree())
plt.title(r'$\sigma_F$ (N=19)')

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```
